```python
# import packages to conduct data analysis and to make interactive charts and maps
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# save file name as a variable with the file name in ""
usa = "USA Raw Data.csv"
am_samoa = "American Samoa Raw Data.csv"
diab = "Diabetes Incidence.csv"
diab_death = "Diabetes Deaths.csv"
# import data as a dataframe to manipulate
df_usa = pd.read_csv(usa)
df_am_samoa = pd.read_csv(am_samoa)
df_diab = pd.read_csv(diab)
df_diab_death = pd.read_csv(diab_death)
```

```python
df_usa.head()
```

```python
df_usa_pivot = pd.pivot_table(df_usa,
                               values = "Value",
                               index = ["Cause of death or injury","Year","Measure"],
                               aggfunc = np.mean).reset_index()

df_am_samoa_pivot = pd.pivot_table(df_am_samoa,
                               values = "Value",
                               index = ["Cause of death or injury","Year","Measure"],
                               aggfunc = np.mean).reset_index()
```

```python
df_usa_pivot
```

```python
df_usa_2017 = df_usa_pivot.query('Year == ["2017"]')
df_usa_2017 = df_usa_2017.sort_values(by ="Value", ascending=False)

df_am_samoa_2017 = df_am_samoa_pivot.query('Year == ["2017"]')
df_am_samoa_2017 = df_am_samoa_2017.sort_values(by ="Value", ascending=False)
```

```python
df_usa_90_17 = df_usa_pivot.query('Year == ["1990 - 2017"]')
df_usa_90_17 = df_usa_90_17.sort_values(by ="Value", ascending=False)

df_am_samoa_90_17 = df_am_samoa_pivot.query('Year == ["1990 - 2017"]')
df_am_samoa_90_17 = df_am_samoa_90_17.sort_values(by ="Value", ascending=False)

```

```python
df_diab_pivot = pd.pivot_table(df_diab,
                               values = "Value",
                               index = ["Location", "Year"],
                               aggfunc = np.mean).reset_index()
df_diab_death_pivot = pd.pivot_table(df_diab_death,
                               values = "Value",
                               index = ["Location", "Year"],
                               aggfunc = np.mean).reset_index()
```

```python
df_diab_am_samoa = df_diab_pivot.query('Location==["American Samoa"]')
df_diab_usa = df_diab_pivot.query('Location==["United States"]')
df_diab_global = df_diab_pivot.query('Location==["Global"]')
```

```python
line_diab_deaths = px.line(df_diab_death_pivot,
                                x = "Year",
                                y = "Value",
                                color = "Location",                                
                                hover_name = "Value",
                                title = "Deaths due to Diabetes per 100,000",
                                labels = {"Year":"Year", "Value":"Deaths"})
line_diab_deaths.write_html("line_diab_deaths.html")
```

```python
line_diab = px.line(df_diab_pivot,
                                x = "Year",
                                y = "Value",
                                color = "Location",                                
                                hover_name = "Value",
                                title = "New Cases of Diabetes per 100,000",
                                labels = {"Year":"Year", "Value":"Cases"})
line_diab.write_html("line_diab.html")
```
